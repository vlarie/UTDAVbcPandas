```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
def priceFormat(result):
    price = "${:,.2f}".format(result)
    return price

uniqueItems = purchase_data["Item ID"].value_counts().count()
avgPrice = priceFormat(purchase_data["Price"].mean())
purchaseCount = purchase_data["Purchase ID"].count()
totalRev = priceFormat(purchase_data["Price"].sum())

summary_table = pd.DataFrame({"Unique Item Count": uniqueItems,
                              "Average Price of Items": avgPrice,
                              "Purchase Count": purchaseCount,
                              "Total Revenue": totalRev}, index=[0])
summary_table
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Appends "Age Group" to original dataframe (purchase_data)
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

# Remove duplicate values for accurate demographics on age groups
ageDF = purchase_data.drop_duplicates(subset="SN")


# Percents
ageDFpercents = ageDF["Age Group"].value_counts(sort=False, normalize=True)

# Counts
ageDFcounts = ageDF["Age Group"].value_counts()


index = ageDFpercents.index.tolist()
cleanAgeDF = pd.DataFrame({"Total Count":ageDFcounts, "Percentage of Players":percentFormat(ageDFpercents)}, index=index)

cleanAgeDF.index.name = "Age Groups"
cleanAgeDF
```
